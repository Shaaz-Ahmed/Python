```python
# Means
```

```python
import pandas as pd

df = pd.read_csv('data.csv')

x = df["Calories"].mean()

df["Calories"].fillna(x)

print(df.to_string())

#As you can see in row 18 and 28, the empty values from "Calories" was replaced with the mean: 304.68
```

```python
import pandas as pd

pf = pd.read_csv('data.csv')

x = pf['Calories'].mean()
pf['Calories'].fillna(x)
print(pf.to_string())
```

```python
# Median
```

```python
# This will first sort in ascending and take the mid value
```

```python
import pandas as pd
import numpy as np
```

```python
pf = pd.read_csv('data.csv')
x = pf['Calories'].median()
pf['Calories'].fillna(x)
print(pf.to_string())
```

```python
# Mode 
```

```python
# Mode gives the most common value
```

```python
import pandas as pd

pf = pd.read_csv('data.csv')

x = pf['Calories'].mode() [0]
pf['Calories'].fillna(x)

print(pf.to_string())
```
