```python
# Labels means we can label the x axis y axis and we can give the title to it
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint)
plt.show()
```

```python
xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint , marker='o' , color='r')
plt.show()
```

```python
xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint , marker='o' , color='r' , linestyle='dotted')
plt.xlabel("This is a X label")
plt.ylabel("This is a Y label")
plt.title("This is a title")
plt.show()
```
