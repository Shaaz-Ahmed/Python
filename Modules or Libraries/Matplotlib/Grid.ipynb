```python
# grid system shows the grid to the graph
# we just need to use grid() function
```

```python
import matplotlib.pyplot as plt
import numpy as np 

xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint)

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint, marker='o')

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint)

plt.grid( axis = 'x')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint)

plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint)

plt.grid(linestyle='dotted')
plt.show()
```
