```python
# Marker in matplotlib
```

```python
# Marker is used to highlight each point of the line 
# we just need to use marker keyword and its value
```

```python
xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint ,marker = 'o' )
plt.show()
```

```python
xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint ,  'o--' )
plt.show()
```
