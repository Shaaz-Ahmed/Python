```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# this is a basic line 
```

```python
xpoint = ([1,2,3,4,5])
ypoint = ([3,7,8,9,10])

plt.plot(xpoint, ypoint)
plt.show()
```

```python
# Line Marker
```

```python
# Line marker means we can give a sign on every point
```

```python
xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint ,marker = 'o')
plt.show()
```

```python
# Line patter
```

```python
# You can keep the line in different different format 
# dotted :
# dashed --
```

```python
xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint ,marker = 'o' , linestyle='dotted')
plt.show()
```

```python
xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint,ypoint ,marker = 'o' , linestyle='dashed')
plt.show()
```
